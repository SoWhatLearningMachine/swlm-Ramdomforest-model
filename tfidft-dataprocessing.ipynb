# Tf-IDf data Parsing, processing

```python
data_tf_idf=pd.read_csv('../input/preprocessed/KNOW_2019_train_tfidf.csv')
```

```python
data_tf_idf_test=pd.read_csv('../input/preprocessed/KNOW_2019_test_tfidf.csv')
```

```python
data_tf_idf.head()
```

```python
data_tf_idf_test.iloc[:,:148].dtypes
```

```python
data_tf_idf_test.iloc[:,148:].dtypes
```

```python
df1=data_tf_idf_test.iloc[:,:148]
df2=data_tf_idf_test.iloc[:,148:]
```

```python
df1=df1.astype(dtype='int32')
```

```python
df2=df2.astype(dtype='float32')
```

```python
data_tf_idf=pd.concat([df1,df2],axis=1)
```

```python
data_tf_idf.to_csv('KNOW_2018_test_tfidf.csv',index=False)
```

## TF-IDF  modeling


```python
X, y = data_tf_idf.drop(["knowcode","index"],axis=1), data_tf_idf.knowcode
```

```python
rfc = RandomForestClassifier(random_state=0,min_samples_leaf=1,min_samples_split=2,n_estimators=100)
scores=cross_val_score(rfc,X,y,scoring='accuracy',cv=5)


print('교차 검증 정확도:', scores)
```

```python
rfc = RandomForestClassifier(random_state=0,min_samples_leaf=1,min_samples_split=2,n_estimators=100)
rfc.fit(X,y)
```

```python
del(X)
del(y)
del(data_tf_idf)
```

```python
X_test, y_test = data_tf_idf_test.drop("index", axis=1), data_tf_idf_test.index
```

```python
data_no_text_test['knowcode'] = rfc.predict(X_test)
data_no_text_test.to_csv('KNOW_2019_test_tfidf.csv',index=False)
```
