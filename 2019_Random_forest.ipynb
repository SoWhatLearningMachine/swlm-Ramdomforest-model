## NO Text

```python
data_no_text=pd.read_csv('../input/preprocessed/KNOW_2019_train_no_text.csv')
data_no_text_test=pd.read_csv('../input/preprocessed/KNOW_2019_test_no_text.csv')
```

```python
X, y = data_no_text.drop(["knowcode","index"],axis=1), data_no_text.knowcode
```

```python
X_test, y_test = data_no_text_test.drop("index", axis=1), data_no_text_test.index
```

```python
from sklearn.model_selection import GridSearchCV

params = { 'n_estimators' : [10, 100],
           'min_samples_leaf' : [1 ,2, 4],
           'min_samples_split' : [2, 4,5,6]
            }

# RandomForestClassifier 객체 생성 후 GridSearchCV 수행
rfc = RandomForestClassifier(random_state = 0, n_jobs = -1)
grid_cv = GridSearchCV(rfc, param_grid = params, cv = 5, n_jobs = -1)
grid_cv.fit(X, y)

print('최적 하이퍼 파라미터: ', grid_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(grid_cv.best_score_))
```

```python
rfc = RandomForestClassifier(random_state=0,min_samples_leaf=1,min_samples_split=4,n_estimators=100)
rfc.fit(X, y)
```

```python
data_no_text_test['knowcode'] = rfc.predict(X_test)
data_no_text_test.to_csv('KNOW_2019_test_no_text.csv',index=False)
```
